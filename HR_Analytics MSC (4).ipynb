# Seperate the dependent and independent columns

```python
x = df.drop("Attrition" , axis = 1)
y = df["Attrition"]
```

# Our data is not balanced first we have to balanced it.

```python
from imblearn.over_sampling import RandomOverSampler
```

# Create the object and fit model on x , y

```python
sample = RandomOverSampler(random_state=42)
x_resampled, y_resampled = sample.fit_resample(x, y)
```

# Now check the data is balanced

```python
pi=y_resampled.value_counts()
```

```python
pi
```

```python
plt.pie(pi, labels=pi.index, autopct='%1.1f%%')
plt.show()
```

# Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.20 , random_state=42)
```

# Model selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Create object of all algorithms

```python
LR = LogisticRegression()
AB = AdaBoostClassifier(n_estimators = 100 , random_state= 42 , learning_rate = 1 )
XG = XGBClassifier()
RF = RandomForestClassifier(n_estimators=100 , max_depth = 4 )
```

# Fit the models 

```python
LR.fit(x_train , y_train)
AB.fit(x_train , y_train)
XG.fit(x_train , y_train)
RF.fit(x_train , y_train)
```

# Predict on the test data

```python
y_pred_LR = LR.predict(x_test)

y_pred_AB = AB.predict(x_test)

y_pred_XG = XG.predict(x_test)

y_pred_RF = RF.predict(x_test)
```

# Print the classification report of all models 

```python
from sklearn.metrics import classification_report
```

```python
print("LogisticRegression")
print(classification_report(y_test , y_pred_LR))
print("---------------------------------------------------")


print("AdaBoostClassifier")
print(classification_report(y_test , y_pred_AB))
print("---------------------------------------------------")



print("XGBClassifier")
print(classification_report(y_test , y_pred_XG))
print("---------------------------------------------------")


print("RandomForestClassifier")
print(classification_report(y_test , y_pred_RF))
print("---------------------------------------------------")
```
