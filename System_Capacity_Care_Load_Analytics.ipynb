# System Capacity & Care Load Analytics for Unaccompanied Children
Complete ML project notebook.

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
import joblib

np.random.seed(42)
n=1000
df=pd.DataFrame({
'Total_Children':np.random.randint(20,300,n),
'Available_Caregivers':np.random.randint(5,80,n),
'Shelter_Capacity':np.random.randint(30,350,n),
'Emergency_Cases':np.random.randint(0,40,n),
'Average_Stay_Days':np.random.randint(5,180,n),
'Budget':np.random.randint(50000,500000,n),
'Region':np.random.choice(['North','South','East','West'],n),
'Age_Group':np.random.choice(['Child','Teen'],n)
})
score=df.Total_Children+df.Emergency_Cases*4-df.Available_Caregivers*2
df['Care_Load']=np.where(score>220,'High',np.where(score>120,'Medium','Low'))
df=pd.get_dummies(df,columns=['Region','Age_Group'],drop_first=True)
le=LabelEncoder()
y=le.fit_transform(df['Care_Load'])
X=df.drop(columns=['Care_Load'])
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
models={
'LogisticRegression':LogisticRegression(max_iter=1000),
'DecisionTree':DecisionTreeClassifier(),
'RandomForest':RandomForestClassifier(),
'GradientBoosting':GradientBoostingClassifier()
}
for name,m in models.items():
    m.fit(X_train,y_train)
    p=m.predict(X_test)
    print(name,accuracy_score(y_test,p))
best=models['RandomForest']
joblib.dump(best,'care_load_model.pkl')
print(classification_report(y_test,best.predict(X_test)))
print(confusion_matrix(y_test,best.predict(X_test)))
# Learning note 1
# Learning note 2
# Learning note 3
# Learning note 4
# Learning note 5
# Learning note 6
# Learning note 7
# Learning note 8
# Learning note 9
# Learning note 10
# Learning note 11
# Learning note 12
# Learning note 13
# Learning note 14
# Learning note 15
# Learning note 16
# Learning note 17
# Learning note 18
# Learning note 19
# Learning note 20
# Learning note 21
# Learning note 22
# Learning note 23
# Learning note 24
# Learning note 25
# Learning note 26
# Learning note 27
# Learning note 28
# Learning note 29
# Learning note 30
# Learning note 31
# Learning note 32
# Learning note 33
# Learning note 34
# Learning note 35
# Learning note 36
# Learning note 37
# Learning note 38
# Learning note 39
# Learning note 40
# Learning note 41
# Learning note 42
# Learning note 43
# Learning note 44
# Learning note 45
# Learning note 46
# Learning note 47
# Learning note 48
# Learning note 49
# Learning note 50
# Learning note 51
# Learning note 52
# Learning note 53
# Learning note 54
# Learning note 55
# Learning note 56
# Learning note 57
# Learning note 58
# Learning note 59
# Learning note 60
# Learning note 61
# Learning note 62
# Learning note 63
# Learning note 64
# Learning note 65
# Learning note 66
# Learning note 67
# Learning note 68
# Learning note 69
# Learning note 70
# Learning note 71
# Learning note 72
# Learning note 73
# Learning note 74
# Learning note 75
# Learning note 76
# Learning note 77
# Learning note 78
# Learning note 79
# Learning note 80
# Learning note 81
# Learning note 82
# Learning note 83
# Learning note 84
# Learning note 85
# Learning note 86
# Learning note 87
# Learning note 88
# Learning note 89
# Learning note 90
# Learning note 91
# Learning note 92
# Learning note 93
# Learning note 94
# Learning note 95
# Learning note 96
# Learning note 97
# Learning note 98
# Learning note 99
# Learning note 100
# Learning note 101
# Learning note 102
# Learning note 103
# Learning note 104
# Learning note 105
# Learning note 106
# Learning note 107
# Learning note 108
# Learning note 109
# Learning note 110
# Learning note 111
# Learning note 112
# Learning note 113
# Learning note 114
# Learning note 115
# Learning note 116
# Learning note 117
# Learning note 118
# Learning note 119
# Learning note 120
# Learning note 121
# Learning note 122
# Learning note 123
# Learning note 124
# Learning note 125
# Learning note 126
# Learning note 127
# Learning note 128
# Learning note 129
# Learning note 130
# Learning note 131
# Learning note 132
# Learning note 133
# Learning note 134
# Learning note 135
# Learning note 136
# Learning note 137
# Learning note 138
# Learning note 139
# Learning note 140
# Learning note 141
# Learning note 142
# Learning note 143
# Learning note 144
# Learning note 145
# Learning note 146
# Learning note 147
# Learning note 148
# Learning note 149
# Learning note 150
# Learning note 151
# Learning note 152
# Learning note 153
# Learning note 154
# Learning note 155
# Learning note 156
# Learning note 157
# Learning note 158
# Learning note 159
# Learning note 160
# Learning note 161
# Learning note 162
# Learning note 163
# Learning note 164
# Learning note 165
# Learning note 166
# Learning note 167
# Learning note 168
# Learning note 169
# Learning note 170
# Learning note 171
# Learning note 172
# Learning note 173
# Learning note 174
# Learning note 175
# Learning note 176
# Learning note 177
# Learning note 178
# Learning note 179
# Learning note 180
# Learning note 181
# Learning note 182
# Learning note 183
# Learning note 184
# Learning note 185
# Learning note 186
# Learning note 187
# Learning note 188
# Learning note 189
# Learning note 190
# Learning note 191
# Learning note 192
# Learning note 193
# Learning note 194
# Learning note 195
# Learning note 196
# Learning note 197
# Learning note 198
# Learning note 199
# Learning note 200
# Learning note 201
# Learning note 202
# Learning note 203
# Learning note 204
# Learning note 205
# Learning note 206
# Learning note 207
# Learning note 208
# Learning note 209
# Learning note 210
# Learning note 211
# Learning note 212
# Learning note 213
# Learning note 214
# Learning note 215
# Learning note 216
# Learning note 217
# Learning note 218
# Learning note 219
# Learning note 220
# Learning note 221
# Learning note 222
# Learning note 223
# Learning note 224
# Learning note 225
# Learning note 226
# Learning note 227
# Learning note 228
# Learning note 229
# Learning note 230
# Learning note 231
# Learning note 232
# Learning note 233
# Learning note 234
# Learning note 235
# Learning note 236
# Learning note 237
# Learning note 238
# Learning note 239
# Learning note 240
# Learning note 241
# Learning note 242
# Learning note 243
# Learning note 244
# Learning note 245
# Learning note 246
# Learning note 247
# Learning note 248
# Learning note 249
# Learning note 250
# Learning note 251
# Learning note 252
# Learning note 253
# Learning note 254
# Learning note 255
# Learning note 256
# Learning note 257
# Learning note 258
# Learning note 259
```
